# import the basic packages

```python
import cv2
from PIL import Image
from IPython.display import display
```

# Displaying the image

```python
image = cv2.imread("LION.jpg")
scale_percent = 25
width = int(image.shape[1]*scale_percent/100)
height = int(image.shape[0]*scale_percent/100)
dsize = (width, height)
image= cv2.resize(image, dsize)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# creating a new image by converting original image int grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# invert the new gray scale image

```python
inverted_image = 255-gray_image
display(Image.fromarray(inverted_image))
```

# Blur the image by using gaussian function in opencv

```python
blurred = cv2.GaussianBlur(inverted_image, (21,21), 0)
display(Image.fromarray(blurred))
```

# Invert the blurred image into pencil sketch

```python
inverted_blurred = 255-blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale = 256.0)
display(Image.fromarray(pencil_sketch))
```

# Original image and Pencil Sketch

```python
display(Image.fromarray(image))
display(Image.fromarray(pencil_sketch))
```
